```python
import matplotlib.pyplot as plt

# Let's define a function that models some business metric, like profit
def calculate_profit(selling_price):
    cost_price = 50
    units_sold = 1000
    return (selling_price - cost_price) * units_sold

selling_prices = list(range(50,101))
profits = []

# Now, let's see what happens if we change the selling price
for price in selling_prices:  # Changed variable here from selling_prices to price
    profit_value = calculate_profit(price)
    print(f"Selling price: {price}, Profit: {profit_value}")
    profits.append(profit_value)
```

```python

# Plot the data
plt.plot(selling_prices, profits)
plt.xlabel('Selling Price')
plt.ylabel('Profit')
plt.title('Profit vs Selling Price')
plt.grid(True)
plt.show();
```

```python
# Lets optimize a more complex relationship

def calc_prof(sp):
  cp = 50
  q = 1000
  return(sp**2-cp) * q # intead of a linear relationship we have a quadratic relationship

sp = list(range(50,101)) # modeling relationship if we increase the price from 50 to 100
p = []

for i in sp:
  p = calc_prof(sp)
  print(f'for price point - {i}, the profit is {p}')
  p.append(p)

# Errors - int object not iterable because you didnt turn sp into a list - list len range no just list range
```

```python
# Monte Carlo Simulation

import numpy as np
import matplotlib.pyplot as plt

# Define a function
def f(x):
    return x**2

# Generate a bunch of random inputs
inputs = np.random.uniform(-1, 1, size=10000)

# Compute the outputs
outputs = f(inputs)

# Now we can see the distribution of the outputs
plt.hist(outputs, bins=50)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Monte Carlo Simulation- Uniform Distribution')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and an axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create some data
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
z = x**2 + y**2

# Plot the data
ax.plot(x, y, z, color='blue')

# Add a title
ax.set_title('3D Surface Plot')

# Add labels to the axes
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Some example data
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([1, 4, 9, 16, 25])

# Fit a quadratic equation to the data
coefficients = np.polyfit(x_data, y_data, 2)

# We can now use this equation to predict y for any x
x_values = np.linspace(0, 6, 100)
y_values = np.polyval(coefficients, x_values)

# Plot the original data and the fit
plt.scatter(x_data, y_data)
plt.plot(x_values, y_values)
plt.show()
```
